- Find the frequencies of sgRNAs in the experiment (just decode) and compare them to the expected frequency (log fold change?)

```python
barcode_df = pd.read_hdf(
    "/home/de64/scratch/de64/sync_folder/2021-11-08_lDE20_Final_3/Barcodes/barcode_output_df.hdf5"
)
```

```python
n_trenches_barcode = barcode_df.groupby("barcodeid")["trenchid"].aggregate(
    lambda x: len(x)
)
n_trenches_sgRNA = barcode_df.groupby("oDEPool7_id")["trenchid"].aggregate(
    lambda x: len(x)
)
```

```python
plt.hist(n_trenches_barcode.values, bins=50, range=(0, 50))
plt.show()
```

```python
plt.hist(n_trenches_sgRNA.values, bins=50, range=(0, 50))
plt.show()
```

```python
sgRNA_df = sgRNA_df.set_index("oDEPool7_id", drop=True)
sgRNA_df["N In Experiment"] = n_trenches_sgRNA
sgRNA_df["N In Experiment"] = (
    sgRNA_df["N In Experiment"].apply(lambda x: 0.0 if np.isnan(x) else x).astype(int)
)
```

```python
plt.hist(sgRNA_df["N In Experiment"].values, bins=50, range=(0, 50))
plt.show()
```

```python
Ntargetid_inExpt = sgRNA_df.groupby("TargetID")["N In Experiment"].sum()
plt.hist(Ntargetid_inExpt, bins=50)
plt.show()
```

```python
sgRNA_df
```

```python
NGene_inExpt = sgRNA_df.groupby("Gene")["N In Experiment"].sum()
plt.hist(NGene_inExpt, bins=50)
plt.show()
```

```python
target_titration_length = sgRNA_df.groupby("TargetID")["N In Experiment"].apply(
    lambda x: np.sum(x >= 5)
)
```

```python
plt.hist(target_titration_length, bins=11)
plt.show()
```

```python
targetid_df = sgRNA_df.groupby("TargetID").apply(lambda x: x.iloc[0])
```

```python
targetid_df["N targetid In Experiment"] = target_titration_length
```

```python
targetid_df["Successful targetids"] = targetid_df["N targetid In Experiment"] >= 4
```

```python
successful_targets_in_gene = targetid_df.groupby("Gene")["Successful targetids"].apply(
    lambda x: np.sum(x)
)
```

```python
plt.hist(successful_targets_in_gene, bins=6)
plt.show()
```

```python
missed_genes = successful_targets_in_gene[
    successful_targets_in_gene == 0
].index.tolist()
```
