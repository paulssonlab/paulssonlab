```python
from scipy.optimize import curve_fit


def exp_fit(row):
    n = 2
    tp = np.arange(0, row.size)
    lb = [0] + (n) * [-1] + n * [0]
    ub = (n + 1) * [1] + n * [1]
    params, _ = curve_fit(
        molco.sum_exp,
        tp,
        row / row[0],
        p0=0.02 * np.ones(2 * n + 1),
        bounds=(lb, ub),
        maxfev=5000,
    )
    return pd.Series(params)


# y = df.filter(regex='^[0-9]+$',axis=1)
# p = y.apply(exp_fit,axis=1)
# k = p[[3,4]]
# k.values.sort(axis=1)
```
