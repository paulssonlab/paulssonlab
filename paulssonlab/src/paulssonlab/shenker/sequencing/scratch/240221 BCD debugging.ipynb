```python
def get_df(alignment):
    return pl.DataFrame({"seq": alignment.split("\n")}).with_columns(
        gaps=pl.col("seq").str.count_matches("-"),
        seq_ungapped=pl.col("seq").str.replace_all("-", ""),
    )
```

```python
df_21x = get_df(alignment_21x)
df_25x = get_df(alignment_25x)
df_29x = get_df(alignment_29x)
df_33x = get_df(alignment_33x)
```

```python
df_25x["gaps"].value_counts(sort=True).with_columns(
    frac=pl.col("count") / pl.col("count").sum()
)
```

```python
margin = 5
counts = df_21x["gaps"].value_counts(sort=True)
correct_gaps = counts[0, 0]
counts.with_columns(
    gap_comp=pl.when(pl.col("gaps") > correct_gaps + margin)
    .then(1)
    .when(pl.col("gaps") < correct_gaps - margin)
    .then(-1)
    .otherwise(0)
).group_by("gap_comp").agg(pl.col("count").sum()).with_columns(
    frac=pl.col("count") / pl.col("count").sum()
)
```
